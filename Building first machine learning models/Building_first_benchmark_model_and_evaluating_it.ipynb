## Steps we will follow to build a model for doing prediction

1. We will begin with dividing the data set into train and valid.
2. Then we will perform mean based and rule based prediction
3. And finally we will evaluate models correctness using MAE

# Step 1 dividing the data

```python
# Importing the train test split to split data
from sklearn.model_selection import train_test_split
```

```python
# Splitting data
pos_train, pos_valid = train_test_split(pos, test_size = 0.2, random_state = 9)
```

```python
# Checking the shape
pos_train.shape, pos_valid.shape
```

# 2. Performing mean based pridiction

```python
# mean of overall unit sold for train data and converting it into integers since units sold can't be integers
unit_sold_mean = int(pos_train['Units_sold'].mean())
unit_sold_mean
```

```python
pos_valid['Pred_Mean_Units_Sold'] = unit_sold_mean
pos_valid.head()
```

# 3. Evaluate Prediction

```python
#Importing mean absolute error
from sklearn.metrics import mean_absolute_error
```

```python
# MAE of overall units sold
mean_absolute_error(pos_valid['Units_sold'],pos_valid['Pred_Mean_Units_Sold'])
```

# 2nd Approach rule based model

```python
#calculate the mean of units sold for pos_train by segment
mean_units_sold_by_segment = pos_train.groupby('Segment')['Units_sold'].mean().astype(int)
mean_units_sold_by_segment
```

```python
# creating a new column pred_Segment_Mean_Units_Sold in the pos_valid dataframe
# and assigining mean

pos_valid['Pred_Segment_Mean_Units_Sold'] = pos_valid["Segment"].map(mean_units_sold_by_segment)
pos_valid.head()
```

# Evaluation

```python
mean_absolute_error(pos_valid["Units_sold"],pos_valid["Pred_Segment_Mean_Units_Sold"])
```
