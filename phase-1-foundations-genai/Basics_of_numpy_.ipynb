```python
#array creation from a python list

list_data = [1, 2, 3, 4, 5]

array_from_list = np.array(list_data)

print(array_from_list)
```

```python
#creating an array of zeros

array_of_zeros = np.zeros((5,2))

print(array_of_zeros)
#
```

```python
#creating an array using range
range_array = np.arange(0,6)

print(range_array)

ranged_skippe_array = np.arange(0,6,2)

print("skipped array: ",ranged_skippe_array)
```

```python
#creating an array of ones

array_of_ones = np.ones((5,2))

print(array_of_ones)

array_of_one = np.ones(6)

print(array_of_one)
```

```python
# creating an array with equal distance

array_of_equal_distance = np.linspace(0,1,5)

print(array_of_equal_distance)
```

```python
#basic indexing

_1darray = np.array([1, 2, 3, 4, 5])

print(_1darray[0])
```

```python
#slicing
# start:0 stop:0 step:0

print(_1darray[0:3])
```

```python
#2d array indeing

arr2d = np.array([[1,2,3],
                  [3,4,5],
                  [6,7,8]])
print(arr2d[1,2])
```

```python
#2d array slicing

print(arr2d[0:2,0:2])
```

```python
#boolean indexing

bool_mask = arr2d <4
print(bool_mask)


print(arr2d[bool_mask])
```

```python
#shape manipulations

matrix = np.arange(1,13)

print(matrix)

matrix = matrix.reshape(3,4)

print(matrix)



#reshape

matrix = matrix.reshape(6,2)
print(matrix)



#ravel

raveled = matrix.ravel()

print(raveled)
```

```python
# matrix transpose

transposed = matrix.transpose()

print(transposed)
```
